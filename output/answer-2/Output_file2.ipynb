```python
import pandas as pd
data_frame = pd.read_csv('https://raw.githubusercontent.com/Bungeetech/internship-test2/master/input/question-2/main.csv')
```

```python
data_frame.head()
```

```python
grouped_df = data_frame.groupby(data_frame['occupation'])
```

```python
max_age = grouped_df.max()
```

```python
max_age = max_age.reset_index()
```

```python
max_age
```

```python
min_age = grouped_df.min()
```

```python
min_age = min_age.reset_index()
```

```python
min_age
```

```python
dictionary2 = {'Occupation':max_age['occupation'], 'min':min_age['age'], 'max':max_age['age']}
```
